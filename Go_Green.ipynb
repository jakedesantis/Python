```python
import matplotlib.pyplot as plt
import numpy as np
import math

theta = np.linspace(0,2*math.pi, 1000)
r = (1+0.9*np.cos(8*theta))*(1+0.1*np.cos(24*theta))*(.9+0.1*np.cos(200*theta))*(1+np.sin(theta))

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color = 'g', linewidth=1)
ax.set_rmax(5.0)
ax.grid(True)
plt.show()
```

```python
rr = 1+np.sin(20*theta)

ax = plt.subplot(111, projection='polar')
ax.plot(theta, rr, color = 'g', linewidth=1)
ax.set_rmax(2.0)
ax.grid(True)
plt.show()
```
